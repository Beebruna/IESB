#### Estrutura de Repetição `while`


1. **Loop enquanto uma condição é verdadeira:**

````python
contador = 0

while contador < 5:
    print(f'Iteração {contador}')
    contador += 1
else:
    print(f'Parei na iteração {contador}')
````

2. **Loop com break para sair com base em uma condição:**

````python
contador = 0
n_parada = 10

while True:
    print(f'Iteração {contador}')
    contador += 1
    if contador >= n_parada:
        print(f'Parei na iteração {contador}')
        break
````

3. **Loop com continue para pular iterações com base em uma condição:**

````python
contador = 0

while contador < 5:
    contador += 1
    if contador == 3:
        continue
    print(f'Iteração {contador}')
````

4. **Loop com else, executado quando a condição do while se torna falsa:**

````python
contador = 0

while contador < 5:
    print(f'Iteração {contador}')
    contador += 1
else:
    print("Loop concluído!")
````

Esses são exemplos que ilustram o uso básico das estruturas de repetição `for` e `while` em Python. Elas são fundamentais para a construção de algoritmos e iterações em seus programas.

Se você precisa modificar uma variável global dentro de uma função, é necessário usar a palavra-chave <code>global</code> para indicar que você está referenciando a variável global e não criando uma nova variável local.

````python
var_global = 100

def func_global():

    var_global += 100
    print(var_global)

func_global()
````

````python
var_global = 100

def func_global():
    global var_global
    var_global += 100
    print(var_global)

func_global()
````

````python
## criando subplots com mosaico

mosaico = """
          CCD
          AAB
          AAB
          """


f, ax = plt.subplot_mosaic(mosaico)

ax['A'].text(x=0.45, y=0.5, s='A', size=20)
ax['B'].text(x=0.45, y=0.5, s='B', size=20)
ax['C'].text(x=0.45, y=0.5, s='C', size=20)
ax['D'].text(x=0.45, y=0.5, s='D', size=20)

print(ax)
````

````python
## criando subplots com mosaico

mosaico = """
          CCD
          AAB
          AAB
          """

f, ax = plt.subplot_mosaic(mosaico)

x,y = np.random.gumbel(size=200), np.random.gumbel(size=200)

ax['A'].scatter(x,y)
ax['B'].hist(y, orientation='horizontal')
ax['C'].hist(x)

ax['A'].sharex(ax['C'])
ax['A'].sharey(ax['B'])

ax['C'].spines['top'].set_visible(False)
ax['C'].spines['right'].set_visible(False)

ax['B'].spines['top'].set_visible(False)
ax['B'].spines['right'].set_visible(False)

ax['D'].axis('off')

print(ax)
````
