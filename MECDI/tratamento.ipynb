```python
# Acessar os valores da série na coluna 'Coluna_A' juntamente com as linhas
for index, value in df['last_name'].items():
    if pd.isna(value):
        print(f'Índice (linha): {index}, Valor: {value}')
        full_name = df.loc[index, 'full_name']
        print(full_name)
        index_unique = valores_unicos.loc[valores_unicos['full_name'] == full_name].index
        print(index_unique)
        if index_unique.empty:
            continue
        last_name = valores_unicos.at[index_unique[0], 'last_name']
        print(last_name)
        df.at[index, 'last_name'] = last_name
```

```python
import matplotlib.pyplot as plt
```

```python
tabela_contingencia = pd.crosstab(df['last_name'], df['year'], margins=True, margins_name='Total')
```

```python
tabela_percentual_colunas = (tabela_contingencia / tabela_contingencia.loc['Total']) * 100
```

```python
tabela_contingencia
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('off')  # Ocultar eixos

# Crie uma tabela a partir do DataFrame e adicione-a à figura
table = ax.table(cellText=tabela_contingencia.values,
                 colLabels=tabela_contingencia.columns,
                 rowLabels=tabela_contingencia.index,
                 loc='center')

table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)
```

```python
plt.savefig('tabela_contingencia.png', bbox_inches='tight')
plt.show()
```
